```python
# separating features and target variable
features = df_crop[['N', 'P','K','temperature', 'humidity', 'ph', 'rainfall']]
target = df_crop['label']
```

```python
# Encode categorical labels into numeric values
label_encoder = LabelEncoder()
target_encoded = label_encoder.fit_transform(target)  # Convert crop names to numbers
```

```python
from sklearn.model_selection import train_test_split
# Split the dataset into training and testing sets
Xtrain, Xtest, Ytrain, Ytest = train_test_split(features, target_encoded, test_size=0.2, random_state=2)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# Initialize and train the QDA classifier
qda = QuadraticDiscriminantAnalysis()
qda.fit(Xtrain, Ytrain)
```

```python
# Predict on the test set
Ypred = qda.predict(Xtest)
```

```python
# Print accuracy
accuracy = accuracy_score(Ytest, Ypred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Classification Report
print("Classification Report:\n", classification_report(Ytest, Ypred))

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(Ytest, Ypred))
```

Quadratic Discriminant Analysis (QDA) model already has 98.64% accuracy

Perform Feature Scaling (Standardization)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
Xtrain_scaled = scaler.fit_transform(Xtrain)
Xtest_scaled = scaler.transform(Xtest)

# Train QDA on scaled data
qda = QuadraticDiscriminantAnalysis()
qda.fit(Xtrain_scaled, Ytrain)

# Predict
Ypred = qda.predict(Xtest_scaled)

# Evaluate
accuracy = accuracy_score(Ytest, Ypred)
print(f"Improved Model Accuracy for QDA: {accuracy * 100:.2f}%")
```

Feature Selection
Not all features contribute equally. You can check feature importance using mutual information.

```python
from sklearn.feature_selection import mutual_info_classif

# Calculate feature importance
importance = mutual_info_classif(Xtrain, Ytrain)
feature_importance = pd.Series(importance, index=features.columns).sort_values(ascending=False)
print(feature_importance)
```

Hyperparameter Tuning
Try different regularization parameters in QDA.

```python
qda = QuadraticDiscriminantAnalysis(reg_param=0.1)  # Adjust regularization
qda.fit(Xtrain_scaled, Ytrain)
Ypred = qda.predict(Xtest_scaled)
print("Tuned Model Accuracy:", accuracy_score(Ytest, Ypred) * 100)
```

Train the PassiveAggressiveClassifier

```python
from sklearn.linear_model import PassiveAggressiveClassifier

pac = PassiveAggressiveClassifier(max_iter=1000, random_state=42)
pac.fit(Xtrain_scaled, Ytrain)

# Predictions
Ypred = pac.predict(Xtest_scaled)

# Evaluate model
accuracy = accuracy_score(Ytest, Ypred)
print(f"PAC Model Accuracy: {accuracy * 100:.2f}%")
print(classification_report(Ytest, Ypred))
```

Hyperparameter Tuning for Better Accuracy

```python

from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'max_iter': [500, 1000, 2000]
}

grid_search = GridSearchCV(PassiveAggressiveClassifier(random_state=42), param_grid, cv=5, scoring='accuracy')
grid_search.fit(Xtrain_scaled, Ytrain)

# Best model
best_pac = grid_search.best_estimator_
Ypred = best_pac.predict(Xtest_scaled)

# Evaluate
print(f"Optimized PAC Model Accuracy: {accuracy_score(Ytest, Ypred) * 100:.2f}%")
print(classification_report(Ytest, Ypred))
```
